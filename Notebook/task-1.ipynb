```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/kaggle/input/datasets/jyotish1628/super-store-dataset/SuperStoreOrders.csv')
```

```python
total_sales = df['sales'].sum()

print("Total Sales:", total_sales)
```

```python
total_profit = df['profit'].sum()

print("Total Profit:", total_profit)
```

```python
#SALES_BY_CATEGORY
category_sales = (
    df.groupby('category')['sales']
    .sum()
    .sort_values()
)

print(category_sales)

plt.figure(figsize=(8,5))
category_sales.plot(kind='barh')

plt.title('Sales by Category')
plt.xlabel('Sales')
plt.ylabel('Category')

plt.tight_layout()
plt.savefig('sales_by_category.png', bbox_inches='tight')
plt.show()
```

```python
#Sales by region
region_sales = (
    df.groupby('region')['sales']
    .sum()
    .sort_values()
)

print(region_sales)

plt.figure(figsize=(10,6))
region_sales.plot(kind='barh')

plt.title('Sales by Region')
plt.xlabel('Sales')
plt.ylabel('Region')

plt.tight_layout()
plt.savefig('sales_by_region.png', bbox_inches='tight')
plt.show()
```

```python
#top_10_products
top_products = (
    df.groupby('product_name')['sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar')

plt.title('Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Sales')

plt.xticks(rotation=90)

plt.tight_layout()
plt.savefig('top_10_products.png', bbox_inches='tight')
plt.show()
```

```python
#revenue trend over time
df['order_date'] = pd.to_datetime(
    df['order_date'],
    dayfirst=True,
    errors='coerce'
)

monthly_sales = (
    df.groupby(df['order_date'].dt.to_period('M'))['sales']
    .sum()
    .sort_index()
)

plt.figure(figsize=(14,6))

monthly_sales.plot()

plt.title('Revenue Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.tight_layout()
plt.savefig('revenue_trend.png', bbox_inches='tight')
plt.show()
```

```python
#profit by category
profit_category = (
    df.groupby('category')['profit']
    .sum()
    .sort_values()
)

print(profit_category)

plt.figure(figsize=(8,5))

profit_category.plot(kind='barh')

plt.title('Profit by Category')
plt.xlabel('Profit')
plt.ylabel('Category')

plt.tight_layout()
plt.savefig('profit_by_category.png', bbox_inches='tight')
plt.show()
```

```python
#TOP MOST PROFITABLE PRODUCTS
top_profit_products = (
    df.groupby('product_name')['profit']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_profit_products)

plt.figure(figsize=(12,6))

top_profit_products.plot(kind='bar')

plt.title('Top 10 Most Profitable Products')
plt.xlabel('Product')
plt.ylabel('Profit')

plt.xticks(rotation=90)

plt.tight_layout()
plt.savefig('top_profit_products.png', bbox_inches='tight')
plt.show()
```
